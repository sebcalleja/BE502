```python
import pandas as pd
import datetime
from datetime import date
```

```python
class Rain:
    def __init__(self, filename1, filename2, gauge_col="gaugeId", date_col="readingDate",
                rain_col="rainAmount", quality_col = "quality"):
        self.gauge_col = gauge_col
        self.date_col = date_col
        self.rain_col = rain_col
        self.quality_col = quality_col
        data1 = pd.read_csv(filename1)
        dates1=data1['readingDate'].str.split('-')
        data1['year']=[dates1[i][0] for i in dates1.index]
        data1['month']=[dates1[i][1] for i in dates1.index]
        data1['day']=[dates1[i][2] for i in dates1.index]
        data2 = pd.read_csv(filename2)
        dates2=data2['readingDate'].str.split('-')
        data2['year']=[dates2[i][0] for i in dates2.index]
        data2['month']=[dates2[i][1] for i in dates2.index]
        data2['day']=[dates2[i][2] for i in dates2.index]
        self.__data1__=data1
        self.__data2__=data2
    def get_rain_t(self, year=None, month=None, day=None, quality=None):
        data=self.__data1__
        if year !=None: data=data[data[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data=data[data[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data=data[data[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data=data[data[self.quality_col].str.upper()==quality.upper()]
        return pd.DataFrame(data[self.rain_col].to_list())
    def get_rain_p(self, year=None, month=None, day=None, quality=None):
        data=self.__data2__
        if year !=None: data=data[data[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data=data[data[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data=data[data[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data=data[data[self.quality_col].str.upper()==quality.upper()]
        return pd.DataFrame(data[self.rain_col].to_list())
    def get_both_tucson(self, year=None, month=None, day=None, quality=None):
        data1=self.__data1__
        data2=self.__data2__
        if year !=None: data1=data1[data1[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data1=data1[data1[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data1=data1[data1[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data1=data1[data1[self.quality_col].str.upper()==quality.upper()]
        if year !=None: data2=data2[data2[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data2=data2[data2[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data2=data2[data2[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data2=data2[data2[self.quality_col].str.upper()==quality.upper()]
        return [data1[self.rain_col], data1[self.gauge_col]]
    def get_both_phoenix(self, year=None, month=None, day=None, quality=None):
        data1=self.__data1__
        data2=self.__data2__
        if year !=None: data1=data1[data1[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data1=data1[data1[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data1=data1[data1[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data1=data1[data1[self.quality_col].str.upper()==quality.upper()]
        if year !=None: data2=data2[data2[self.date_col].str.contains(str(year)+"-")] 
        if month!=None and year!=None:
           month_str="%02d" %month
           data2=data2[data2[self.date_col].str.contains(str(year)+"-"+month_str+"-")]
        if day!=None and month!=None and year!=None:
           day_str="%02d" %day
           data2=data2[data2[self.date_col].str.contains(str(year)+"-"+month_str+"-"+day_str)]
        if quality!=None: data2=data2[data2[self.quality_col].str.upper()==quality.upper()]
        return [data2[self.rain_col], data2[self.gauge_col]]
rain_both = Rain('tucson_rain.csv', 'phoenix_rain.csv')
x=pd.DataFrame(rain_both.get_both_tucson(2019,1,1,quality="good"))
y=pd.DataFrame(rain_both.get_both_phoenix(2019,1,1,quality="good"))
```
